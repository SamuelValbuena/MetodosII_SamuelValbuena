```python
import numpy as np
import matplotlib.pyplot as plt
def func_complex(z):
    return 1j * z
def runge_kutta_3rd_order_complex(z0, h, steps):
    z = np.zeros(steps+1, dtype=complex)
    z[0] = z0
    for i in range(steps):
        k1 = func_complex(z[i])
        k2 = func_complex(z[i] + 0.5 * h * k1)
        k3 = func_complex(z[i] - h * k1 + 2 * h * k2)
        z[i+1] = z[i] + (h / 6) * (k1 + 4 * k2 + k3)
    return z
def runge_kutta_4th_order_complex(z0, h, steps):
    z = np.zeros(steps+1, dtype=complex)
    z[0] = z0
    for i in range(steps):
        k1 = func_complex(z[i])
        k2 = func_complex(z[i] + 0.5 * h * k1)
        k3 = func_complex(z[i] + 0.5 * h * k2)
        k4 = func_complex(z[i] + h * k3)
        z[i+1] = z[i] + (h / 6) * (k1 + 2*k2 + 2*k3 + k4)
    return z
z0 = 1 + 0j  
h = 0.1      
steps = 100  
z_3rd_order = runge_kutta_3rd_order_complex(z0, h, steps)
z_4th_order = runge_kutta_4th_order_complex(z0, h, steps)
plt.figure(figsize=(8, 8))
plt.plot(z_3rd_order.real, z_3rd_order.imag, label='3rd Order Runge-Kutta', marker='o')
plt.plot(z_4th_order.real, z_4th_order.imag, label='4th Order Runge-Kutta', marker='x')
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Runge-Kutta Methods Stability in the Complex Plane')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def riccati_diff_eq(x, y):
    return x**4 * y**2 - 2 * x**2 * y - 1 / x**3
def analytic_solution(x):
    return 1 / (-1 - x**2 / 2) + 1 / x**2
def euler_method(x, y, h, diff_eq):
    for i in range(1, len(x)):
        y[i] = y[i-1] + h * diff_eq(x[i-1], y[i-1])
x_values = np.linspace(np.sqrt(2), 10, 1000)
h = x_values[1] - x_values[0]
y_numerical = np.zeros_like(x_values)
y_numerical[0] = 0
euler_method(x_values, y_numerical, h, riccati_diff_eq)
y_analytic = analytic_solution(x_values)
plt.plot(x_values, y_numerical, label="Analiticamente")
plt.plot(x_values, y_analytic, label="Numericamente")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def du_dt(u, q):
    return u**q
def euler_method(q):
    t0 = 0
    tf = 10
    dt = 0.01  
    num_steps = int((tf - t0) / dt) + 1
    t_values = np.linspace(t0, tf, num_steps)
    u_values = np.zeros_like(t_values)
    u_values[0] = 1 
    for i in range(1, num_steps):
        u_values[i] = u_values[i-1] + dt * du_dt(u_values[i-1], q)
    return t_values, u_values
q_values = [0., 0.2, 0.4, 0.7, 0.9, 1.]
for q in q_values:
    t, u = euler_method(q)
    plt.plot(t, u, label=f"q = {q}")
t_values = np.linspace(0, 10, 1000)
plt.plot(t_values, np.exp(t_values), label="q = 1 (exact)")
for q in q_values[1:]:
    if 1 - q > 0:
        plt.plot(t_values, (t_values*(1 - q) + 1)**(1/(1-q)), label=f"q = {q} (exact)")
plt.xlabel("t")
plt.ylabel("u(t)")
plt.title("Solución numérica y analítica de la ecuación diferencial")
plt.legend()
plt.grid(True)
plt.show()
```
